### 3.2. Phoning it in üìû [3]

Standardize the format of the ```'phone_number'``` column by removing any non-numeric characters and ensuring that all phone numbers have the same length.

````python

# Defining a function to standardize the phone numbers in phone_number column
def clean_phoneNumber(phoneNumber):
    # Filter to remove non-numeric characters
    phoneNumber = ''.join(filter(str.isdigit, phoneNumber))
    
    # Condition to check all phone numbers has 10 length
    if len(phoneNumber) == 10:
        return phoneNumber
    else:
        return None
# Applying the function to the 'phone_number' column
df['phone_number'] = df['phone_number'].apply(clean_phoneNumber)    
````

### 3.3. No more HTML üìÑ [2]
Find columns containing HTML tags and replace them with an appropriate plain text equivalent, such as a newline character or space.

ex:

```html
<p>
  Feast on delicious grub at Jerry's Famous Deli.<br />
  Its retro-style casual setting features comfortable booth seating.
</p>
```

to:

```
Feast on delicious grub at Jerry's Famous Deli. Its retro-style casual setting features comfortable booth seating.
```

````python

# AT FIRST, FINDING OUT COLUMNS HAVING HTML TAGS 
# Defining a function to check whether a string contains HTML tags
def Containing_any_html_tags(text):
    if isinstance(text, str):
        return any(tag in text for tag in ['<', '>'])
    return False

# To find out columns containing HTML tags
html_columns = [col for col in df.columns if df[col].apply(Containing_any_html_tags).any()]

# To display the names of columns containing HTML tags
print("Columns containing HTML tags:", html_columns)
````

````python
from bs4 import BeautifulSoup

# Define a function to clean HTML tags from a single string
def Clean_html_tags(text):
    soup = BeautifulSoup(text, 'html.parser')
    # To replace HTML tags with spaces or newlines
    cleaned_text = soup.get_text(separator=" ")
    return cleaned_text
````

````python
# Applying the function to required columns
df['description'] = df['description'].apply(Clean_html_tags)
df['public_transit'] = df['public_transit'].apply(Clean_html_tags)
df['cross_street'] = df['cross_street'].apply(Clean_html_tags)
df['entertainment'] = df['entertainment'].apply(Clean_html_tags)
df['operating_hours'] = df['operating_hours'].apply(Clean_html_tags)
df['most_recent_review'] = df['most_recent_review'].apply(Clean_html_tags)
````

Below, you've been provided with some seed words to help you in analysing each aspect of the restaurant's experience. Feel free to modify it as you see fit.

````python
seed_words = {
    'Food': {
        'Positive': ['tasty', 'delicious', 'flavorful', 'fresh', 'juicy', 'crispy', 'savory', 'yummy', 'delectable', 'succulent'],
        'Negative': ['bland', 'stale', 'overcooked', 'undercooked', 'greasy', 'salty', 'burnt', 'sour', 'tasteless', 'dry']
    },
    'Service': {
        'Positive': ['friendly', 'attentive', 'prompt', 'courteous', 'professional', 'helpful', 'accommodating', 'quick', 'efficient', 'warm'],
        'Negative': ['slow', 'rude', 'inattentive', 'disorganized', 'negligent', 'unprofessional', 'unhelpful', 'snobby', 'aloof', 'cold']
    },
    'Ambiance': {
        'Positive': ['cozy', 'inviting', 'elegant', 'trendy', 'clean', 'lively', 'comfortable', 'romantic', 'vibrant', 'charming'],
        'Negative': ['loud', 'crowded', 'dirty', 'dull', 'dark', 'cramped', 'noisy', 'uncomfortable', 'drab', 'claustrophobic']
    }
}
````

#### Sentiment Analysis üé≠ [5]

**Perform a sentiment scoring (`Negative` or `Positive`) of each mention of the aspects.**

> The seed words dictionary already accounts for the sentiment of each aspect, so feel free to utilize it once more.

````python
# Defining a function to perform sentiment scoring
def sentiment_scoring(review_text):
    sentiment_score = 0

    for aspect, sentiment_words in seed_words.items():
        for sentiment, words in sentiment_words.items():
            for word in words:
                if word in review_text:
                    if sentiment == 'Positive':
                        sentiment_score += 1
                    else:
                        sentiment_score -= 1

    return 'Positive' if sentiment_score > 0 else 'Negative'

# Apply sentiment scoring to the review text
reviews_df['sentiment'] = reviews_df['review_text'].apply(sentiment_scoring)
````

````python

# Print the DataFrame with sentiment scores
print(reviews_df[['brand_name', 'sentiment']])
````

#### Compile üìö [5]

**Provide an analysis of the aspect and sentiment of restaurants in Los Angeles. Provide at least 1 visualization and summarize your findings.**

````python
# Initializing sentiment scores and sentiment labels
sentiment_scores = {aspect: [] for aspect in seed_words}
sentiment_labels = {aspect: [] for aspect in seed_words}

# Filter reviews for Los Angeles
los_angeles_reviews = reviews_df[reviews_df['city'] == 'Los Angeles']

# Function to calculate sentiment scores and labels
def calculate_sentiment(review_text):
    scores = {aspect: 0 for aspect in seed_words}
    labels = {aspect: '' for aspect in seed_words}
    words = review_text.split()
    
    for aspect, sentiment_words in seed_words.items():
        for sentiment, seed_word_list in sentiment_words.items():
            for word in words:
                if word.lower() in seed_word_list:
                    scores[aspect] += 1 if sentiment == 'Positive' else -1
    for aspect in seed_words:
        labels[aspect] = 'Positive' if all(score >= 0 for score in sentiment_scores[aspect]) else 'Negative'
    
    return scores, labels

# Apply sentiment scoring and labels to each review
for index, row in los_angeles_reviews.iterrows():
    scores, labels = calculate_sentiment(row['review_text'])
    for aspect in seed_words:
        sentiment_scores[aspect].append(scores[aspect])
        sentiment_labels[aspect].append(labels[aspect])

# Add sentiment scores and labels to the DataFrame
for aspect in seed_words:
    los_angeles_reviews[f'{aspect}_Scores'] = sentiment_scores[aspect]
    los_angeles_reviews[f'{aspect}_Labels'] = sentiment_labels[aspect]
````
